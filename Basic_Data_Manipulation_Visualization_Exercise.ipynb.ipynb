###  28. Import the dataset occupation.csv and assign object as users

```python
users = pd.read_csv('occupation.csv',sep = '|')
users
```

### 29. Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].mean().reset_index()
```

### 30. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
users['gender_encoded'] = np.where(users['gender'] == 'M', 1, 0)
male_ratio_per_occupation = users.groupby('occupation')['gender_encoded'].mean().sort_values(ascending=False)
male_ratio_per_occupation
```

### 31. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].agg(['min', 'max']).reset_index()
```

### 32. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender'])['age'].mean().reset_index()
```

### 33.  For each occupation present the percentage of women and men

```python
occupation_counts = users.groupby('occupation')['gender_encoded'].value_counts().unstack().fillna(0)

# Calculate percentages
occupation_counts['Total Users'] = occupation_counts.sum(axis=1)
occupation_counts['Percentage Men'] = (occupation_counts[0] / occupation_counts['Total Users']) * 100
occupation_counts['Percentage Women'] = (occupation_counts[1] / occupation_counts['Total Users']) * 100

occupation_counts
```
